```python
generator = Generate_Graph(graph_type="OW", 
                           nb_vertices= V,
                           dropout_edge_rate=dropout_edge_rate)

g, adj_matrix, all_edges = generator.build()
```

```python
n_sto = 8
n_adv = 0
n_hyb = 4
agents_sto = ['sto{}'.format(i) for i in range(1,n_sto+1)]
agents_adv = ["adv{}".format(i) for i in range(1,n_adv+1)]
agents_hyb = ["hyb{}".format(i) for i in range(1,n_hyb+1)]
agents = agents_sto + agents_adv + agents_hyb

summary_experiences = run_experiment(graph_type="OW", number_vertices=20, dropout_edge_rate=0.9,
                                         nb_iterations=500,
                                         n_sto=n_sto, n_adv=n_adv, n_hyb=n_hyb, 
                                         n_rand=0, n_fix=0,
                                         exploration_parameter_stoc=5, eta_hyb=0.1, Bt_hyb=0.01, use_adaptif_lr_hyb=False,
                                         gamma_hyb=10.0, time_limit=500)
```

```python
def draw_network(A, start=None, destination=None, path=None):
    
    colors = ["b"] * len(A)
    colors[start] = "r"
    colors[destination] = "g"
    for p in path[1:-1]:
        colors[p] = "y"
        
    G = nx.from_numpy_matrix(np.matrix(A), create_using=nx.DiGraph)
    layout = nx.spring_layout(G, scale =10, seed=10)
    labels = nx.get_edge_attributes(G, "weight")
    
    nx.draw(G, layout, with_labels = True, node_size = 500, node_color=colors)
    nx.draw_networkx_edge_labels(G, pos=layout, edge_labels=labels);
def draw_networks(A,
                  summary_experiences, 
                  agents,
                  timestep,
                  ):
    rows = int(len(agents)/3)
    fig, ax = plt.subplots( nrows=rows+1,
                            ncols=3 ,
                            figsize=(15, 15))
    plt.suptitle(f"iteration {timestep}", y=1.05)
    summary = summary_experiences[timestep]
    for i in range(len(summary)):
        sum_ = summary[i]
        plt.subplot(rows+1, 3, i+1)
        plt.title(f"agent {agents[i]} | Cost = {sum_['cost']}")
        start = sum_["path"][0]
        destination = sum_["path"][-1]
        path = sum_["path"]
        draw_network(A=A,
                     start=start,
                     destination= destination,
                     path=path,
                    )
    
    plt.tight_layout()
```

```python
draw_networks(A=adj_matrix,
              summary_experiences=summary_experiences,
              agents=agents,
              timestep=0)
```
